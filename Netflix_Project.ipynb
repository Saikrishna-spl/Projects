```python
import pandas as pd #importing the data
df=pd.read_csv("https://d2beiqkhq929f0.cloudfront.net/public_assets/assets/000/000/940/original/netflix.csv") #Imported CSV file
df.head(10) # displaying top 10 rows
```

# **Basic Analysis**

**Handling Null Values**

```python
df.fillna(
    {'director':'unknown director','cast':'unknown actor','country':'unknown country','date_added':'unknown date','rating':'unknown','duration':'unknown'}
    ,inplace=True) # Filling the catergorical columns with unknown suffix
df
```

# **Un-nesting the columns**
The list of columns needed to unnest are **director**,**cast**,**country** and **listed in**.

**Un-nesting the director column**

```python
df_dir=df.loc[:,['title','director']] # Creating a new data frame
df_dir=df_dir[df_dir['director']!='unknown director'] # Dropping unknown directors
split_dir=df_dir['director'].str.split(',',expand=True) #Splitting the listed directors
df_dir=pd.concat([df_dir,split_dir],axis=1) #Joining both dataframes
df_dir=pd.melt(df_dir,id_vars=['title'],value_vars=split_dir.columns,value_name='directors',var_name='grouped_directors') # converting columns into rows
df_dir=df_dir.drop('grouped_directors',axis=1) # Dropping grouped_directors column
df_dir=df_dir.dropna(subset=['directors']) #Dropping the null values
df_dir=df_dir.drop_duplicates() #Dropping the duplicates
df_dir=df_dir.reset_index(drop=True) #Re-Indexing
df_dir
```

**Un-nesting the cast column**

```python
df_cast=df.loc[:,['title','cast']] # Creating a new data frame
df_cast=df_cast[df_cast['cast']!='unknown actor'] #Dropping Unknown actor
split_cast=df_cast['cast'].str.split(',', expand=True) #Splitting the listed cast
df_cast = pd.concat([df_cast, split_cast], axis=1) #Joining both dataframes
df_cast = pd.melt(df_cast, id_vars=['title'], value_vars=split_cast.columns, value_name='actor') # converting columns into rows
df_cast=df_cast.drop(['variable'],axis=1) # Dropping variable column
df_cast=df_cast.dropna(subset=['actor']) #Dropping the null values
df_cast=df_cast.drop_duplicates() #Dropping the duplicates
df_cast=df_cast.reset_index(drop=True) #Re-Indexing
df_dir
df_cast
```

**Un-nesting the country column**

```python
df_country=df.loc[:,['title','country']] # Creating a new data frame
df_country=df_country[df['country']!='unknown country'] #Dropping Unknown country
split_country=df_country['country'].str.split(',',expand=True)  #Splitting the listed countries
df_country=pd.concat([df_country,split_country],axis=1) #Joining both dataframes
df_country=pd.melt(df_country,id_vars='title',value_vars=split_country.columns,value_name='country_name') # converting columns into rows
df_country=df_country.drop('variable',axis=1) # Dropping variable column
df_country=df_country.dropna(subset=['country_name'])
df_country=df_country.drop_duplicates()
df_country=df_country.reset_index(drop=True) #Re-Indexing
df_country
```

**Un-nesting the Genre in listed_in column**

```python
df_genre=df.loc[:,['title','listed_in']] # Creating a new data frame
split_genre=df_genre['listed_in'].str.split(',',expand=True) #Splitting the listed genre
df_genre=pd.concat([df_genre,split_genre],axis=1) #Joining both dataframes
df_genre=pd.melt(df_genre,id_vars='title',value_vars=split_genre.columns,value_name='genre') # converting columns into rows
df_genre=df_genre.drop('variable',axis=1)  # Dropping variable column
df_genre=df_genre.dropna(subset=['genre'])
df_genre=df_genre.drop_duplicates()
df_genre=df_genre.reset_index(drop=True) #Re-Indexing
df_genre
```

**Merging df_dir,df_country,df_cast,df_genre and remaining columns from df into a new data frame**

```python
df_dc=pd.merge(df_dir,df_country,on='title',how='left') # Merging directors and country dataframes into
df_dcC=pd.merge(df_dc,df_cast,on='title',how='left') # Merging cast data frame with previous
df_dcCg=pd.merge(df_dcC,df_genre,on='title',how='left') #Merging genre with previos data frame
df_remaining=df.loc[:,['title','show_id','type','date_added','release_year','rating','duration','description']] #Extraction of remaining columns from original data frame
df_final=pd.merge(df_dcCg,df_remaining,on='title',how='left') # Merging all the columns from original that are not in previous one
```

**Creating month and week from the date_added column**

```python
df_final['month']=pd.to_datetime(df['date_added'],errors='coerce').dt.month_name() #Extraction of month
df_final['week']=pd.to_datetime(df['date_added'],errors='coerce').dt.isocalendar().week # Extraction of week

```

**Replacing null values**

```python
df_final.fillna({'country_name':'unknown country','actor':'unknown actor','month':'unknown month','week':0},inplace=True)
df_final
```

# **Find After how many days the movie will be added to Netflix after the release of the movie**

```python
release_year_new=pd.to_datetime(df['release_year'].astype(str) + '-01-01', errors='coerce') # adding month and date to year and converting it to date_time
date_added_new=pd.to_datetime(df_final['date_added'],errors='coerce') # converting date_added into date_time format
number_of_days= (date_added_new-release_year_new).mode() # Finding difference between the dates in days
number_of_days
```

**Conclusion**: a movie will be added to Netflix after **609** days the release of the movie
